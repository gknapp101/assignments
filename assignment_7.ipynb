## 8.
- Open the METABRIC data. Make a histogram of 'Ratio Therapy'.
- Let treatment, $T$ be distributed binomial with parameter $p$. Then the contribution to the likelihood for each patient $i$, with $y_i = 0$ for no radiation therapy and $y_i=1$ for radiation therapy, is 
$$
p^{y_i}(1-p)^{1-y_i}
$$
- Write out the likelihood.
- Maximize the likelihood with respect to $p$. What is the MLE, $\hat{p}$?
- Bootstrap the sampling density/distribution of $\hat{p}$.

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy as sp 

df = pd.read_csv('metabric.csv')

sns.histplot(df['Radio Therapy'],stat='proportion', color='orange')
```

2. Likelihood
$$
L(p) = \prod_{i=1}^n p^{y_i}(1-p)^{1-y_i}
$$

3. Maximize Likelihood/MLE

$$
L(p) = \prod_{i=1}^n p^{y_i} (1-p)^{1-y_i}
$$

$$
\ell(p) = \log L(p) = \sum_{i=1}^n \big[ y_i \log p + (1-y_i) \log(1-p) \big]
$$

$$
\frac{d\ell}{dp} = \sum_{i=1}^n \frac{y_i}{p} - \sum_{i=1}^n \frac{1-y_i}{1-p} = 0
$$

$$
\frac{\sum y_i}{p} = \frac{n - \sum y_i}{1-p}
$$

$$
\sum y_i (1-p) = (n - \sum y_i) p
$$

$$
\sum y_i = n p
$$

$$
\hat{p} = \frac{\sum_{i=1}^n y_i}{n}
$$

MLE =  $$\hat{p} = \frac{\sum_{i=1}^n y_i}{n}$$


```python
# Basis of Code taken from assignment 7 solutions.ipynb

# Convert YES/NO to numeric 1/0 for the bootstrap
y = pd.to_numeric(df['Radio Therapy'].replace({'NO': '0', 'YES': '1'}), errors='coerce')

# Number of bootstrap resamples
S = 5000

# Bootstrap distribution of the sample mean
sample = [np.mean(y.sample(frac=1, replace=True)) for s in range(S)]

# Create side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# KDE of the bootstrap sample means
sns.kdeplot(x=sample, ax=axes[1], label='kde', fill=True, color='orange')
axes[1].set_title('Sample KDE')

# ECDF of the bootstrap sample means
sns.ecdfplot(x=sample, ax=axes[0], label='ecdf', color='orange')
axes[0].set_title('Sample ECDF')

plt.show()
```

2. Density
$$
Y \sim N(\mu,\sigma^2)
$$
$$
X = e^{Y}.
$$

$$
Y = \ln X, \qquad x > 0.
$$

$$
f_Y(y) = \frac{1}{\sigma \sqrt{2\pi}}
\exp\!\left( -\frac{(y-\mu)^2}{2\sigma^2} \right).
$$

$$
f_X(x) = f_Y(\ln x)\left|\frac{d}{dx}\ln x\right|.
$$


$$
\frac{d}{dx}\ln x = \frac{1}{x}.
$$


$$
f_X(x)
= \frac{1}{x}\cdot \frac{1}{\sigma\sqrt{2\pi}}
\exp\!\left( -\frac{(\ln x - \mu)^2}{2\sigma^2} \right), \qquad x>0.
$$


$$
\boxed{
f_X(x)
= \frac{1}{x\,\sigma\sqrt{2\pi}}
\exp\!\left( -\frac{(\ln x - \mu)^2}{2\sigma^2} \right),
\quad x>0
}
$$

Steps:

1. Start with a normal variable $(Y \sim N(\mu,\sigma^2))$.
2. Define the new variable $(X = e^Y), so (Y = \ln X)$.
3. Use the change-of-variables formula for transformations:
   density of (X) = density of (Y) evaluated at $(\ln x)  derivative of (\ln x)$.
4. Compute the derivative: $(d(\ln x)/dx = 1/x)$.
5. Multiply the normal density $(f_Y(\ln x)) by (1/x)$.
6. This produces the log-normal PDF.


$$
f(y) = \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{y-\mu}{\sigma} \right)^2}
$$
<br>
<br>
3. Likelihood

$$
L(\mu, \sigma) = \prod_{i=1}^n \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{y_i-\mu}{\sigma} \right)^2}.
$$
<br><br>
4. maximize the likelihood
$$
L(\mu, \sigma) = \prod_{i=1}^n \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{y_i-\mu}{\sigma} \right)^2}.
$$
$$
\ell(\mu, \sigma) = \sum_{i=1}^n \log \left( \dfrac{1}{\sqrt{2\pi}}\right) - \log(\sigma) -\frac{1}{2}\left( \frac{y_i-\mu}{\sigma} \right)^2
$$
$$
\ell(\mu, \sigma) = -\frac{n}{2}\log(2\pi) - n\log(\sigma) - \frac{1}{2\sigma^2} \sum_{i=1}^n (y_i - \mu)^2
$$

$$
\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2} \sum_{i=1}^n (y_i - \mu) = 0 \quad \Rightarrow \quad \hat{\mu} = \frac{1}{n} \sum_{i=1}^n y_i
$$
$$
\frac{\partial \ell}{\partial \sigma} = -\frac{n}{\sigma} + \frac{1}{\sigma^3} \sum_{i=1}^n (y_i - \mu)^2 = 0 \quad \Rightarrow \quad \hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{\mu})^2
$$

5. MLE
$$
\hat{\mu} = \frac{1}{n} \sum_{i=1}^n y_i
$$

$$
 \hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{\mu})^2
$$


```python
# Basis of Code taken from assignment 7 solutions.ipynb

# Estimate parameters of the log-normal distribution
# Take log of 'price' and compute mean and standard deviation
mu_hat = np.mean(np.log(df['price']))
sigma_hat = np.sqrt(np.mean((y - mu_hat)**2))

# Grid of values for plotting PDF and CDF
grid = np.linspace(9, 14, 100)

# PDF and CDF of the normal distribution (for the log-transformed variable)
pdf = sp.stats.norm.pdf(grid, loc=mu_hat, scale=sigma_hat)
cdf = sp.stats.norm.cdf(grid, loc=mu_hat, scale=sigma_hat)

# Plot KDE vs PDF and ECDF vs CDF side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# KDE plot of the data with PDF overlay
sns.kdeplot(x=y, ax=axes[0], label='kde')
sns.lineplot(x=grid, y=pdf, ax=axes[0], label='pdf', color='red')
axes[0].set_title('KDE vs PDF')

# ECDF plot of the data with CDF overlay
sns.ecdfplot(x=y, ax=axes[1], label='ecdf')
sns.lineplot(x=grid, y=cdf, ax=axes[1], label='cdf', color='red')
axes[1].set_title('ECDF vs CDF')

plt.show()
```

The ECDFs line up extremely closely  the bootstrap curve essentially lies directly on top of the original, indicating an excellent fit. For the PDF, the bootstrap distribution shows a slightly sharper peak and a hint of left-skew, but it still retains a broadly bell-shaped form with tails that taper off quickly. Overall, the resampled distribution closely mirrors the shape of the original.


```python
# Estimation function using Maximum Likelihood Estimation (MLE)
# Returns estimates of mean and standard deviation for a sample
def mle(y):
    mu_hat = np.mean(y)  # MLE of the mean
    sigma_hat = np.sqrt(np.mean((y - mu_hat)**2))  # MLE of the standard deviation
    return mu_hat, sigma_hat

# Bootstrapping to generate sampling distributions
S = 5000  # Number of bootstrap samples
sample = [mle(y.sample(frac=1.0, replace=True)) for s in range(S)]

# Extract bootstrap estimates for the mean
mu_hat_sample = [row[0] for row in sample]

# Plot KDE and ECDF of bootstrap mean estimates
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.kdeplot(x=mu_hat_sample, ax=axes[0], label='kde', color='red', fill=True)
axes[0].set_title('Sampling Density, $\\hat{\\mu}$')
sns.ecdfplot(x=mu_hat_sample, ax=axes[1], label='ecdf', color='red')
axes[1].set_title('Sampling Distribution, $\\hat{\\mu}$')
plt.show()

# Extract bootstrap estimates for the standard deviation
sigma_hat_samples = [row[1] for row in sample]

# Plot KDE and ECDF of bootstrap standard deviation estimates
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.kdeplot(x=sigma_hat_samples, ax=axes[0], label='kde', color='red', fill=True)
axes[0].set_title('$\\hat{\\sigma}$, Sampling Density')
sns.ecdfplot(x=sigma_hat_samples, ax=axes[1], label='ecdf', color='red')
axes[1].set_title('$\\hat{\\sigma}$, Sampling Distribution')
plt.show()
```

- Open the METABRIC data. Make a histogram of 'Mutation Count' with around 50 bins. Let $Y$ be the mutation count the random variable, and $y_i$ the mutation count for patient $i$.
- We're going to model this using the Poisson Distribution. The probability that $y_i = k$ is given by
$$
pr[y_i = k ] = \frac{ \lambda^k e^{-\lambda} }{k!}
$$
- Write out the likelihood.
- Maximize the likelihood. What is the MLE? (Hint: The sample mean. As usual.)
- Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.
- Bootstrap the sampling density/distribution of your parameter.

```python
df = pd.read_csv('metabric.csv')
y = df['Mutation Count']
sns.histplot(y,stat='proportion', bins=50)
```

2. Likelihood



$$L(\lambda) = \prod_{i=1}^{n} \frac{\lambda^{y_i} e^{-\lambda}}{y_i!}$$

3. Maximize the Likelihood

$$L(\lambda) = \prod_{i=1}^{n} \frac{\lambda^{y_i} e^{-\lambda}}{y_i!}$$

$$\ell(\lambda) = \sum_{i=1}^{n} y_i \log(\lambda) - n\lambda - \sum_{i=1}^{n} \log(y_i!)$$

$$\frac{d\ell(\lambda)}{d\lambda} = \frac{\sum_{i=1}^{n} y_i}{\lambda} - n = 0$$

   **MLE (the sample mean):** $$\hat{\lambda} = \frac{1}{n} \sum_{i=1}^{n} y_i = \bar{y}$$



```python

# Estimate Poisson rate (lambda) using sample mean
lambda_hat = np.mean(y)

# Define grid of x-values for plotting PMF and CDF
grid = np.arange(0, int(np.max(y)) + 5)

# Poisson PMF and CDF based on estimated lambda
pmf = sp.stats.poisson.pmf(grid, mu=lambda_hat)
cdf = sp.stats.poisson.cdf(grid, mu=lambda_hat)

# Create side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# PMF comparison with histogram of data
axes[0].hist(y, bins=grid, density=True, alpha=0.6, label='empirical', edgecolor='black')
axes[0].plot(grid, pmf, label='pmf', color='red')
axes[0].set_title('KDE vs PMF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('probability')
axes[0].legend()

# CDF comparison with ECDF of data
sns.ecdfplot(x=y, ax=axes[1], label='ecdf')
axes[1].plot(grid, cdf, label='cdf', color='red')
axes[1].set_title('CDF vs ECDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cumulative probability')
axes[1].legend()

plt.show()
```

The fitted distribution captures the overall shape of the empirical data fairly well, with the mode of the fitted PMF lining up closely with the peak of the histogram and the lower-count region showing a good match. However, the right tail reveals a clear deviation: the empirical data exhibit more large values than the model predicts, causing the fitted PMF to drop off more quickly than the observed frequencies. This mismatch indicates that the fitted distribution underestimates tail probabilities.

The CDF comparison tells a similar story. For small to moderate values of x, the fitted CDF follows the ECDF closely, but in the middle range the ECDF shifts slightly to the right, suggesting that probability accumulates a bit later in the empirical data. At higher values, the ECDF rises more slowly than the fitted CDF, reinforcing that the empirical distribution contains more extreme values than expected under the model.
